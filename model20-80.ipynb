```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
malware_file = "/home/eder/projects/cybersecurity/data/GrupoC/malware_features.csv"
goodware_file = "/home/eder/projects/cybersecurity/data/GrupoI/goodware_features.csv"
malware = pd.read_csv(malware_file)
goodware = pd.read_csv(goodware_file)
print(malware.head())
print(goodware.head())
```

```python
(malware.columns == goodware.columns).all()
```

```python
malware.shape, goodware.shape
```

```python
malware_label = np.ones(malware.shape[0])
goodware_label = np.zeros(goodware.shape[0])
```

```python
malware["label"] = malware_label
goodware["label"] = goodware_label
dataset = pd.concat([malware, goodware], ignore_index=True)
dataset.head()
```

```python
train_data, test_data = train_test_split(dataset, test_size=0.1, random_state=42, stratify=dataset["label"])
#x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42, stratify=y_train)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor = TabularPredictor(label="label").fit(train_data)
```

```python
# to load best model: predictor = TabularPredictor.load("/home/eder/projects/cybersecurity/AutogluonModels/ag-20260403_083643")
predictor.fit_summary()
```

```python
predictor.evaluate_predictions(y_true=test_data["label"], y_pred=predictor.predict(test_data))
```
